```python
import pandas as pd
df = pd.read_csv('raw_data.csv', sep=',', header=None, index_col=None)
lookup_table = pd.read_csv('lookup.csv')
```

```python
clean_data = pd.DataFrame({})

for col in df:
    survey_num = list(df[col])[-1]
    order = list(lookup_table[survey_num])
    classes = list(df[col][16:-1])
    data = list(df[col][0:16])
    sorted_data = [0]*16
    for idx, val in enumerate(order):
        sorted_data[val-1] = data[idx]
    sorted_data.extend(classes)
    clean_data = pd.concat([clean_data, pd.Series(sorted_data)], axis=1)
    clean_data = clean_data.rename(columns={0:"%s" %col})


clean_data.to_csv('clean_data.csv', index=False)
clean_data
```
